```python
import pandas as pd
#Load the dataset
df = pd.read_csv('World_pop_total.csv')
print(df.columns)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MultipleLocator
# Filtered dataset for the year 2022
data = {
    'Country Name': ['Nigeria'],
    'Population, Female': [108093075],
    'Population, Male': [110448136]
}
# Create the DataFrame
df = pd.DataFrame(data)


# Function to format the y-axis labels in millions
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x * 1e-6)

# Create the formatter
formatter = FuncFormatter(millions)

# Plotting the bar chart
plt.figure(figsize=(8, 6))

# Bar positions and width
positions = range(len(df))
width = 0.01

# Plot Male Population
plt.bar([p + width for p in positions], df['Population, Male'], width=width, label='Male', color='r')

# Plot Female Population
plt.bar([p - width/2 for p in positions], df['Population, Female'], width=width, label='Female', color='b')

# Set the y-axis formatter to display population in millions
plt.gca().yaxis.set_major_formatter(formatter)

# Set the y-axis locator to intervals of 10 million
plt.gca().yaxis.set_major_locator(MultipleLocator(10000000))

# Set the title and labels
plt.title('Population Distribution by Gender in 2022')
plt.xlabel('Nigeria')
plt.ylabel('Population')
plt.legend()

# Show the plot
plt.show()
```
